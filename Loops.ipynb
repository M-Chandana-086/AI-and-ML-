```python
i=0
while i<=10:
  print('hello')
  i=i+1
```

```python
i=10
while i>=1:
  print(i*2)
  i-=1
```

```python
#print 10 to 170 but only the 13 multiples
for i in range(10,171):
  if i%13==0:
    print(i)
```

```python
i=10
while i<=170:
  if i%13==0:
    print(i)
  i+=1
```

```python
'''
*
*
*
*
*
'''
for i in range(5):
  print('*')
```

```python
i=0
while i<=5:
  print('*',end=' ')
  i+=1
```

```python
'''
*****
*****
'''
for i in range(1,6):
  for j in range(4):
    print('*',end=' ')
  print('\n')
```

```python
'''
*
**
***
****
'''
for i in range(1,6):
  print('*'*i)
```
