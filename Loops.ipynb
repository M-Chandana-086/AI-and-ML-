```python
#find a factorial of a num(multiplication of consecutive numbers)
n=int(input('enter a num'))
fact=1
while n:#true
  fact=fact*n #5*4=20
  n=n-1
print(fact)
```

```python
#check whether a given num is prime or not
n=int(input('enter a num'))
if n<=1:
  print('not prime')
else:
  is_prime=True#assume it is prime
  for i in range(2,n):
    if n%i==0:
      is_prime=False
      break
  if is_prime:
    print('prime')
  else:
    print('not prime')
```
